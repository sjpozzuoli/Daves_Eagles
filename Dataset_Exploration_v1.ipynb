```python
#import dependencies
import pandas as pd
```

```python
#load in Realtor.com dataset sourced from kaggle | RDC_Inventory_Hotness_Metrics_County_History.csv
#https://www.kaggle.com/maedemaftouni/real-estate-market-trends?select=RDC_Inventory_Hotness_Metrics_County_History.csv

file_to_load = "Realtor_Kaggle_Dataset.csv"
realtor_df = pd.read_csv(file_to_load)

realtor_df.head()
```

```python
realtor_df.dtypes
```

```python
#Select the columns - our team has decided to focus on year over year data to try and account for seasonality 
# and to explore trends in cities on the rise
cleaned_df = realtor_df[["month_date_yyyymm","cbsa_title","nielsen_hh_rank","hotness_rank","hotness_rank_yy","supply_score",
                        "demand_score","median_days_on_market_yy","ldp_unique_viewers_per_property_yy",
                        "ldp_unique_viewers_per_property_vs_us","median_listing_price_yy","quality_flag"]]
cleaned_df.head()
```

```python
#Examine rows where the quality flag has been triggered 
triggered_df = cleaned_df.loc[cleaned_df["quality_flag"] == 1]
triggered_df.head()
```

```python
#The number of rows that trip the quality flag is
print(len(triggered_df["quality_flag"]))
```

```python
#The total length of the clean df is
print(len(cleaned_df["quality_flag"]))
```

```python
#Given the number of rows is relatively small compared to the overall dataset size, drop the potential inaccurate rows
cleaned_df = cleaned_df.loc[cleaned_df["quality_flag"] != 1]
print(len(cleaned_df["quality_flag"]))
```

```python
#Drop the quality_flag column 
cleaned_df = cleaned_df.drop(["quality_flag"], axis=1)
```

```python
#Examine data for null values
cleaned_df.isnull().sum()
```

```python
#Check for alternative columns, given the high number of year over year null values
realtor_df.isnull().sum()
```

```python
#Creating a save point for dropping null values using the year over year columns
year_over_year_df = cleaned_df.dropna()
print(len(year_over_year_df["cbsa_title"]))
```

```python
#create a new cleaned df to explore alternative column options
cleaned_df = realtor_df[["month_date_yyyymm","cbsa_title","nielsen_hh_rank","hotness_rank","supply_score",
                        "demand_score","median_days_on_market","median_listing_price","quality_flag"]]
```

```python
#Assess the number of rows that trip the quality flag
triggered_df = cleaned_df.loc[cleaned_df["quality_flag"] == 1]
print(len(triggered_df["quality_flag"]))
```

```python
#Assess the number of null values in the new cleaned DF
cleaned_df.isnull().sum()
```

```python
#Drop the triggered rows from the new clean_df 
cleaned_df = cleaned_df.loc[cleaned_df["quality_flag"] != 1]
print(len(cleaned_df["quality_flag"]))
```

```python
#Drop the quality_flag column and the null values with rows without a cbsa_title (eg. city name)
cleaned_df = cleaned_df.drop(["quality_flag"], axis=1)
cleaned_df = cleaned_df.dropna()
print(len(cleaned_df["cbsa_title"]))
```

```python
#Change the first column to a date time type
cleaned_df["month_date_yyyymm"] = pd.to_datetime(cleaned_df["month_date_yyyymm"],format='%Y%m')
cleaned_df
```

```python
#Change the first column to a date time type
year_over_year_df["month_date_yyyymm"] = pd.to_datetime(year_over_year_df["month_date_yyyymm"],format='%Y%m')
```

```python
#Split the cbsa_title column into cities and state columns
title_split = cleaned_df["cbsa_title"].str.split(", ",n=1, expand=True)
title_split
```

```python
#Split the cities and state columns into cities and state columns based on the "-" delimiter to ensure 
#there's only 1 value in each in the cleaned_df using the current values 
cities_split = title_split[0].str.split("-",n=-1, expand=True)
cities_split.head()
```

```python
state_split = title_split[1].str.split("-",n=-1, expand=True)
state_split.head()
```

```python
#Add the first column of each newly created split dataframe back to the cleaned_df and drop the cbsa_title column
cleaned_df["city_name"] = cities_split[0]
cleaned_df["state"] = state_split[0]
cleaned_df = cleaned_df.drop(["cbsa_title"], axis=1)
cleaned_df.head()
```

```python
#Repeat the splitting process on the YoY dataframe
title_split = year_over_year_df["cbsa_title"].str.split(", ",n=1, expand=True)

cities_split = title_split[0].str.split("-",n=-1, expand=True)
state_split = title_split[1].str.split("-",n=-1, expand=True)

year_over_year_df["city_name"] = cities_split[0]
year_over_year_df["state"] = state_split[0]
year_over_year_df = year_over_year_df.drop(["cbsa_title"], axis=1)
year_over_year_df.head()
```

```python
# Create the output file (CSV) for the cleaned_df.
output_data_file = "static_data.csv"
# Export the City_Data into a CSV.
cleaned_df.to_csv(output_data_file, index=False)
```

```python
# Create the output file (CSV) for the yoy_df.
output_data_file = "year_over_year_data.csv"
# Export the City_Data into a CSV.
year_over_year_df.to_csv(output_data_file, index=False)
```
