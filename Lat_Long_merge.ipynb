```python
import pandas as pd
```

```python
cities_to_load = "Data_Main/âˆž"
cities_df = pd.read_csv(cities_to_load)

cities_df.head()
```

```python
clusters_to_load = "Data_Main/data_clusters_ready.csv"
clusters_df = pd.read_csv(clusters_to_load)

clusters_df.head()
```

```python
cities_df_cleaned = cities_df[['STATE_CODE', 'CITY', 'LATITUDE', 'LONGITUDE']]
cities_df_cleaned
```

```python
cities_df_cleaned['city_state'] = cities_df_cleaned['CITY'].str.cat(cities_df_cleaned ['STATE_CODE'],sep=", ")
cities_df_cleaned
```

```python
cities_df_cleaned = cities_df_cleaned[['city_state', 'LATITUDE', 'LONGITUDE']]
cities_df_cleaned['city_state'] = cities_df_cleaned['city_state'].str.lower()
cities_df_cleaned = cities_df_cleaned.drop_duplicates(subset='city_state', keep='first')
cities_df_cleaned
```

```python
clusters_df['city_state'] = clusters_df['city'].str.cat(clusters_df ['state'],sep=", ")
clusters_df
```

```python
merged_df = pd.merge(clusters_df, cities_df_cleaned, on="city_state", how='left')
merged_df
```

```python
#clusters_latlong_df = merged_df.drop(['city_state'],axis=1)
clusters_latlong_df = merged_df.rename(columns={'LATITUDE': 'latitude', 'LONGITUDE': 'longitude'})
clusters_latlong_df
```

```python
clusters_latlong_july21_df = clusters_latlong_df.loc[clusters_latlong_df["date"] == "2021-07-01"]
clusters_latlong_july21_df
```

```python
clusters_latlong_df = clusters_latlong_df.dropna()
clusters_latlong_july21_df = clusters_latlong_july21_df.dropna()
clusters_latlong_july21_df = clusters_latlong_df.drop_duplicates(subset='city_state', keep='first')
clusters_latlong_july21_df
```

```python
#clusters_latlong_df = clusters_latlong_df.rename(columns={'LATITUDE': 'latitude', 'LONGITUDE': 'longitude'})
clusters_latlong_df.to_csv('Data_Main/data_clusters_latlong_ready.csv', index=False)
clusters_latlong_july21_df.to_csv('Data_Main/data_clusters_latlong_july21.csv', index=False)
```
